 üëâ                     **Z={...,‚àí3,‚àí2,‚àí1,0,1,2,3,...}**

```python
print (10)
print (type(10))

# como declarar un numero entero
variable = 100
print (variable)
print (type(variable))

# se puede declarar usando la funcion "int"
variable_2 = int(20.0)
print (variable_2)
print (type(variable_2))
```

# N√∫meros con punto flotante
## Floating - float

- Son n√∫meros que tienen parte decimal
- Pueden ser positivos, negativos o cero.
- Tienen precisi√≥n limitada, no pueden representar n√∫meros con una cantidad infinita de decimales.

 **R={...,‚àí2.5,‚àí1.5,‚àí0.5,0.5,1.5,2.5,...}**

```python
# Valor 0.5 Flotante
print (0.5)
print (type(0.5))

# DECLARAR UN VALOR FLOTANTE
variable_4 = 0.100546
print(variable_4)
print (type (variable_4))

variable_7 = float (1)
print(variable_7)
print (type (variable_7))

# LIMITE DE PRECISION FLOTANTE
# Precision de 17 decimales
variable_5 = 0.999999999999999
print(variable_5)
print (type (variable_5))
```

# Operadores aritm√©ticos
Son s√≠mbolos especiales que representan operaciones matem√°ticas como la suma, la multiplicaci√≥n y otros c√°lculos
## Operadores aritm√©ticos en Python
- +: Suma
- -: Resta
- ‚àó : Multiplicaci√≥n
- / : Divisi√≥n
- ‚àó‚àó : Potencia
- // : Divisi√≥n entera
- % : M√≥dulo o residuo

```python
# Operadores aritmeticos
a = 10
b = 3
# Suma
print ("suma")
print (a + b)
# Resta
print ("resta")
print (a - b)
# Multiplicacion 
print ("multiplicacion")
print (a * b)
# Division
print ("division")
print (a / b)
# Potencia
print ("Potencia")
print (a ** b)
# Modulo o residuo
print ("modulo o residuo")
print (a % b)
# Division entera
print ("Division entera")
print (a // b)
```

# Operadores de comparaci√≥n

ü§îSon s√≠mbolos especiales que permiten comparar dos valores.

## Operadores de comparaci√≥n en Python

- <: Estrictamente menor que
- >: Estrictamente mayor que
- ==: Estrictamente igual que
- <=: Menor o igual que
- >=: Mayor o igual que
- != : Estrictamente diferente que

```python
print ("operadores de comparacion con int - float")
entero = 12
flotante = 12.1
print (entero < flotante)
print (entero > flotante)
print (entero == flotante)
print (entero <= flotante)
print (entero >= flotante)
print (entero != flotante)
```
