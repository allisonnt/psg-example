## Ejercicio 02: Crear una imagen con símbolos usando print

**Descripción:**

Este ejercicio consiste en utilizar la función `print()` de Python para dibujar
una figura o imagen simple en la consola. La imagen se formará mediante la
combinación de diferentes caracteres y símbolos.

**Objetivo:**

- Practicar el uso de la función `print()` para salida de texto.
- Desarrollar la creatividad para representar formas visuales con caracteres.
- Entender cómo múltiples llamadas a `print()` pueden construir una imagen línea por línea.

**Ejemplo de imagen a crear:** Un pequeño cohete. 🐍

```python
# COHETE PYTHON 🐍

print("      /\\")
print("     /  \\")
print("    /____\\")
print("   |      |")
print("   |      |")
print("   |  PY  |") # PY por Python :)
print("   |      |")
print("  /|------|\\")
print(" / |      | \\")
print("/  `------'  \\")
print("   /______\\")
print("     //||\\\\")
print("    // || \\\\")
print("   //__||__\\\\")
print("      ==== ")
print("     (____)")
```
